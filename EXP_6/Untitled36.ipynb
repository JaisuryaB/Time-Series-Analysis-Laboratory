```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Step 3: Load the dataset
df = pd.read_csv('gold_data (1).csv')

# Step 4: Convert 'Date' to datetime and sort
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df = df.sort_values('Date')

# Step 5: Set 'Date' as index for time series operations
df.set_index('Date', inplace=True)

# Step 6: Apply Moving Average Smoothing
window_size = 7  # You can change this to experiment
df['SMA_Price'] = df['Price'].rolling(window=window_size).mean()

plt.figure(figsize=(14, 6))
plt.plot(df['Price'], label='Original Price', alpha=0.5)
plt.plot(df['SMA_Price'], label=f'{window_size}-Day SMA', linewidth=2, color='orange')
plt.title('Gold Price with Moving Average Smoothing')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from google.colab import files

# Step 2: Upload your CSV file
uploaded = files.upload()

# Step 3: Load the dataset
df = pd.read_csv('gold_data (1).csv')  # Make sure the filename matches

# Step 4: Convert 'Date' to datetime and sort
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df = df.sort_values('Date')

# Step 5: Set 'Date' as index
df.set_index('Date', inplace=True)

# Step 6: Perform time series decomposition
decomposition = seasonal_decompose(df['Price'], model='additive', period=365)

# Step 7: Plot the decomposition
fig = decomposition.plot()
fig.set_size_inches(14, 10)
plt.suptitle("Time Series Decomposition (Additive)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
